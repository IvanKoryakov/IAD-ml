**Задание 12**

```python
print('Средний балл за экзамен по чтению для каждой этнической группы:')
start = time.time()
dict = {}
names = df['race/ethnicity'].unique() # получаем список различных групп
for group in sorted(names):
  subset = df[df['race/ethnicity'] == group] # выделяем группу отдельно от других
  average = subset['reading score'].mean() # находим среднее значение для выделенной группы
  dict[group] = '{:.6f}'.format(average) # округляем до 6 знаков после запятой + добавляем в словарь
for name in dict.keys():
  print(name, '  ', dict[name])
end = time.time() - start
print('\nВремя выполнения:', end)

print('\nМинимальный балл за экзамен по письму для каждого уровня образования:')

start = time.time()
dict = {}
names = df['education'].unique()  # получаем список различных типов образования
for type in sorted(names):
  subset = df[df['education'] == type] # выделяем тип отдельно от других
  average = subset['writing score'].min() # находим минимальное значение для выделенной группы
  dict[type] = average # добавляем в словарь
for name in dict.keys():
  print(name, '  ', dict[name])
end = time.time() - start
print('\nВремя выполнения:', end)
```

**Задание 14**

```python
df['Math_PassStatus'] = np.where(df['math score'] < passmark, 'F', 'P')
print('Сдали экзамен по математике:', len(df[df['Math_PassStatus'] == 'P']))
print('Не сдали экзамен по математике:', len(df[df['Math_PassStatus'] == 'F']))

df['Read_PassStatus'] = np.where(df['reading score'] < passmark, 'F', 'P')
print('\nСдали экзамен по чтению:', len(df[df['Read_PassStatus'] == 'P']))
print('Не сдали экзамен по чтению:', len(df[df['Read_PassStatus'] == 'F']))

df['Writ_PassStatus'] = np.where(df['writing score'] < passmark, 'F', 'P')
print('\nСдали экзамен по письму:', len(df[df['Writ_PassStatus'] == 'P']))
print('Не сдали экзамен по письму:', len(df[df['Writ_PassStatus'] == 'F']))
```

**Задание 16**

```python
def GetGrade(average_mark):
  if average_mark < 50:
    return 'F'
  if average_mark >= 50 and average_mark < 60:
    return 'E'
  if average_mark >= 60 and average_mark < 70:
    return 'D'
  if average_mark >= 70 and average_mark < 80:
    return 'C'
  if average_mark >= 80 and average_mark <= 90:
    return 'B'
  if average_mark > 90:
    return 'A'
df['Grade'] = (df['math score'] + df['reading score'] + df['writing score'])/3
df['Grade'] = df['Grade'].apply(GetGrade)
print(df['Grade'].value_counts().sort_index())
```
